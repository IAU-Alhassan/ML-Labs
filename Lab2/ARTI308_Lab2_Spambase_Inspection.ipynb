**Dataset:** Spambase (UCI Machine Learning Repository)  
**Problem Type:** Binary Classification  
**Target variable:** `class` (0 = non-spam / legitimate, 1 = spam)  
**Goal:** Predict whether an email is spam or not based on word frequencies, character frequencies, and capital letter statistics.

```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_44_spambase.csv")
```

```python
print("Shape of the dataset:", df.shape)
```

```python
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\nColumns and data types:")
print(df.dtypes)
```

```python
print("\nClass distribution (0 = non-spam, 1 = spam):")
print(df['class'].value_counts())
print("\nPercentage:")
print(df['class'].value_counts(normalize=True) * 100)
```
